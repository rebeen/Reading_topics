```python
import pandas as pd
import numpy as np
```

```python
lista=pd.read_csv('topic1.txt', sep="\t", header=None)[0].tolist() 

```

```python
lista
```

```python
condistion_list=[]
for item in lista[0::2]:
    for sub_item in item.split(" "):
        condistion_list.append(sub_item)
        
condistion_list=[x for x in condistion_list if x]
print(condistion_list)
```

```python
value_list=[]
for item in lista[1::2]:
    for sub_item in item.split(" "):
        value_list.append(sub_item)
        
value_list=[x for x in value_list if x]
print(value_list)
```

```python
{key:value for key, value in zip(condistion_list,value_list)}
```
